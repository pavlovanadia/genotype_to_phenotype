```python
import pandas as pd
```

```python
human_mouse_genes = pd.read_csv("../databases/HMD_HumanPhenotype.rpt", delimiter="\t", header=None)\
    .rename(columns={0: "gene_human", 1: "entrez_id_human", 2: "gene_mouse", 3: "MGI", 4: "MP"})
```

```python
human_mouse_genes.MP.notna().sum()
```

```python
human_mouse_genes = human_mouse_genes.drop(['MP', 5], axis=1)
```

```python
human_mouse_genes
```

```python
mouse_gene_phenotype = pd.read_csv("../databases/MGI_Geno_DiseaseDO.rpt.txt", delimiter="\t", header=None)
```

```python
mouse_gene_phenotype = mouse_gene_phenotype.drop([0, 1, 2, 3, 5, 7, 8], axis=1).rename(columns={4: "MP", 6: "MGI"})
```

```python
mouse_gene_phenotype
```

```python
mouse_g_to_p_wide = mouse_gene_phenotype.groupby("MGI").agg({"MP": lambda x: list(x)})
```

```python
mouse_g_to_p_wide
```

```python
with_MP_long = pd.merge(human_mouse_genes, mouse_gene_phenotype, how='left', on = 'MGI').drop_duplicates()
```

```python
with_MP_wide = pd.merge(human_mouse_genes, mouse_g_to_p_wide, how='left', on = 'MGI')
```

```python
with_MP_wide.head()
```

```python
with_MP_wide.MP.notna().sum()
```

```python
with_MP_wide.to_csv("current_no_human.csv", sep='\t')
```
