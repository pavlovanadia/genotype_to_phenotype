```python
import pandas as pd
```

```python
human_part = pd.read_csv("db_3_human.csv", delimiter='\t')
mouse_part = pd.read_csv("general_bd_with_mp.tsv", delimiter='\t')
```

```python
human_part
```

```python
mouse_part
```

```python
check_mouse_pheno = pd.merge(human_part, mouse_part, on="MGI", how="left")
check_mouse_pheno = check_mouse_pheno[["MGI", "MP_x", "MP_y"]]
```

```python
check_mouse_pheno["len_x"] = check_mouse_pheno.MP_x.apply(lambda x: len(str(x).split(", ")))
check_mouse_pheno["len_y"] = check_mouse_pheno.MP_y.apply(lambda x: len(str(x).split(",")))
```

```python
check_mouse_pheno["different"] = check_mouse_pheno.len_x > check_mouse_pheno.len_y
```

```python
check_mouse_pheno[check_mouse_pheno.different == True]
```

```python
merged_hum_mouse = pd.merge(human_part, mouse_part, on="MGI", how="left")
merged_hum_mouse = merged_hum_mouse\
    .drop(['Unnamed: 0.1', 'Unnamed: 0_x', 'MP_x', 'Unnamed: 0_y'], axis=1)\
    .rename(columns={"MP_y": "MP"})
```

```python
merged_hum_mouse.to_csv("db_4_entrez_hp_mgi_mp.tsv", sep='\t')
```
